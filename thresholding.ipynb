# Image Thresholding

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

image = cv2.imread('PreprocessedDatabase\\alphonso\\front\\alphonso_front_n9oxq9caa4i7k83zauda.jpg')
img2rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # Convert from BGR to RGB color
img2gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Convert to Grayscale


plt.axis('off')
plt.title('Grayscale Image', fontsize=20)
plt.imshow(img2gray, cmap="gray")
```

## Color Channels

```python
# Red channel
r_channel = img2rgb[:, :, 0]
# Green channel
g_channel = img2rgb[:, :, 1]
# Blue channel
b_channel = img2rgb[:, :, 2]

# Plot Original Image
plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(img2rgb)

# Plot Red Channel Image
plt.subplot(2, 2, 2)
plt.title('Red channel')
plt.axis('off')
plt.imshow(r_channel, cmap='gray')

# Plot Green Channel Image
plt.subplot(2, 2, 3)
plt.title('Green Channel')
plt.axis('off')
plt.imshow(g_channel, cmap='gray')

# Plot Blue Channel Image
plt.subplot(2, 2, 4)
plt.title('Blue Channel')
plt.axis('off')
plt.imshow(b_channel, cmap='gray')

plt.show()
```

## Gray Scale Histogram

```python

# Plot Grayscale Image
plt.subplot(2, 1, 1)
plt.title('Gray Scale Image')
plt.axis('off')
plt.imshow(img2gray, cmap='gray')

# Plot Grayscale Histogram
plt.subplot(2, 1, 2)
plt.title('Grayscale Histogram')
plt.xlabel('Grayscale intensities')
plt.ylabel('Frequency')
plt.hist(img2gray.ravel(), bins=256, range=(0, 256))
plt.show()
```

### Otsu Threshold Value

```python
retval, thresh_img = cv2.threshold(img2gray, 0, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('Threshold Value : {0}'.format(retval)) # Print threshold value returned by Otsu Thresholding Algorithm
plt.title('Binary Threshold Image',fontsize=20)
plt.axis('off')
plt.imshow(thresh_img, cmap='gray')
plt.show()
```

### Manual Threshold Value

```python
retval, thresh_img = cv2.threshold(img2gray, 90, 255,cv2.THRESH_BINARY_INV)
print('Threshold Value : {0}'.format(retval))
plt.title('Thresholded Image')
plt.axis('off')
plt.imshow(thresh_img, cmap='gray')
plt.show()
```

```python
plt.title('Threshold Histogram')
plt.xlabel('Intensities')
plt.ylabel('Frequencies')
plt.hist(thresh_img.ravel(), bins=256, range=(0, 256))
plt.show()
```
