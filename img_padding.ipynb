# Script to pad dataset images
### Requirements - Images should have preprocessed and shadows should have removed.

```python
import os
from PIL import Image
from matplotlib import pyplot as plt
from utils import get_file_paths

%matplotlib inline
```

```python
BOX_WIDTH = 700
BOX_HEIGHT = 400

def pad_image(pil_img):
    width, height = pil_img.size
    v_pad = (BOX_HEIGHT - height)
    h_pad = (BOX_WIDTH - width)
    padding_top = v_pad // 2
    padding_right = h_pad // 2
    padding_bottom = v_pad // 2
    padding_left = h_pad // 2
    result = Image.new(pil_img.mode, (BOX_WIDTH, BOX_HEIGHT), (0,0,0))
    result.paste(pil_img, (padding_left, padding_top))
    return result
```

```python
img_file = "C:\\Users\\Asus\\Desktop\\Research\\PreprocessedDatabase\\alphonso\\back\\alphonso_back_16_p.jpg"
plt.axis("off")
padded = pad_image(Image.open(img_file))
plt.title("Padded Image Size : (%d, %d)" % padded.size)
plt.imshow(padded)
```

## Pad data set images in loop

```python
DATASET_DIR = "PreprocessedDatabase"
WORKING_DIR = os.getcwd()
IMAGES_DIR_PATH = os.path.join(WORKING_DIR, DATASET_DIR)

image_files = get_file_paths(IMAGES_DIR_PATH, extension=['.jpg'], recursive=True)

for img_file in image_files:
    img = Image.open(img_file)
    padded = pad_image(img)
    padded.save(img_file)

print("Successfully Padded %d Images" % len(image_files))
```
